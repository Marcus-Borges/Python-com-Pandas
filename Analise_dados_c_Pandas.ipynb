```python
# Carregado arquivo - basecensus.csv
dados2 = pd.read_csv('basecensus.csv')
```

```python
# Comando Dataframe.shape e Dataframe.describe()
print(dados2.shape)
# Mostra o número de linhas e colunas (linhas, colunas)

# Só as linhas
print(dados2.shape[0])
```

```python
display(dados2.head(16))
```

```python
# Quais são as colunas e que tipo de dados cada um tem.
print(dados2.info())
```

```python
# describe() mostra as estatisticas "mais básicas" que a gente precisa
print(dados2.describe())
```

```python
# Filtrar apenas por uma coluna
display(dados2['education'])
```

```python
# Criando um novo dataframe com colunas mais específicas.
dados3 = dados2[['age', 'workclass', 'occupation', 'sex']]
display(dados3)
```

```python
# Utiliznado o describe() para mostrar média de idade ('age') no df criado anteriormente ('dados3')
print(dados3.describe())
```

```python
dados4 = dados3[['age', 'sex']].groupby(by='sex').mean().round(2)
display(dados4)
```
